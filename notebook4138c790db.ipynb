```python
import pandas as pd

df = pd.read_csv("/kaggle/input/competitions/titanic/train.csv")

df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df['Age'].hist()
plt.title("Age Distribution")
plt.show()

# Boxplot
sns.boxplot(x=df['Fare'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.pairplot(df[['Age','Fare','Survived']])
plt.show()
```

# Observations:

1. Younger passengers had slightly higher survival.
2. Higher fare passengers survived more.
3. Fare contains outliers.
4. Age distribution is right skewed.
5. Survival is correlated with passenger class.
